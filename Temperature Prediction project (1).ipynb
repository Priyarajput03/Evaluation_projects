## This is a Regression Problem since Target/ Label columns ('Next_Tmax','Next_Tmin') have Continuous type of Data.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor, GradientBoostingRegressor
from sklearn.neighbors import KNeighborsRegressor
from xgboost import XGBRegressor
from sklearn.svm import SVR
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.model_selection import cross_val_score
```

## Model Building

```python
max_acc = 0
max_rs = 0

for i in range(200):
    x_train,x_test,y_train,y_test = train_test_split(X_scaled,y,test_size = 0.20,random_state = i)
    LR = LinearRegression()
    LR.fit(x_train,y_train)
    pred = LR.predict(x_test)
    acc = r2_score(y_test,pred)
    if acc>max_acc:
        max_acc = acc
        max_rs = i
print('Best accuracy score is',max_acc,'on random state',max_rs)
```

```python
def reg(model,X_scaled,y):
    x_train,x_test,y_train,y_test=train_test_split(X_scaled,y,test_size=0.20,random_state=128)
    
    model.fit(x_train,y_train) #trainnng the model
    
    pred=model.predict(x_test) 
    
    rmse=mean_squared_error(y_test,pred)  # findthe root mean squared error value
    print('RMSE is:',rmse)
    
    r2=r2_score(y_test,pred)*100 # finding r2 score value
    print('r2 score is:',r2)

    cvs=(cross_val_score(model,X_scaled,y,cv=5).mean())*100 # cross validation score
    print('Cross validation score is:', cvs)
    
    result=r2-cvs
    print('R2 score -Cross validation score is:', result)
    
```

```python
model=LinearRegression()
reg(model,X_scaled,y)
```

```python
#Lasso Regressor
from sklearn.linear_model import Lasso, Ridge
model=Lasso(alpha=0.01,normalize=True)
reg(model,X_scaled,y)
```

```python
#Ridge Regressor
model=Ridge(alpha=0.01,normalize=True)
reg(model,X_scaled,y)
```

```python
model=RandomForestRegressor()
reg(model,X_scaled,y)
```

```python
model=DecisionTreeRegressor()
reg(model,X_scaled,y)
```

```python
model=KNeighborsRegressor()
reg(model,X_scaled,y)
```
