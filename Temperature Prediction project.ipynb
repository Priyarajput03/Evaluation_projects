```python
from sklearn.preprocessing import PowerTransformer #importing the library

scaler = PowerTransformer(method='yeo-johnson')
    
df['LDAPS_RHmax'] = scaler.fit_transform(df['LDAPS_RHmax'].values.reshape(-1,1))
df['LDAPS_Tmax_lapse'] = scaler.fit_transform(df['LDAPS_Tmax_lapse'].values.reshape(-1,1))
df['LDAPS_WS'] = scaler.fit_transform(df['LDAPS_WS'].values.reshape(-1,1))
df['LDAPS_LH'] = scaler.fit_transform(df['LDAPS_LH'].values.reshape(-1,1))

df['LDAPS_CC1'] = scaler.fit_transform(df['LDAPS_CC1'].values.reshape(-1,1))
df['LDAPS_CC3'] = scaler.fit_transform(df['LDAPS_CC3'].values.reshape(-1,1))
df['LDAPS_CC4'] = scaler.fit_transform(df['LDAPS_CC4'].values.reshape(-1,1))

df['LDAPS_PPT1'] = scaler.fit_transform(df['LDAPS_PPT1'].values.reshape(-1,1))
df['LDAPS_PPT2'] = scaler.fit_transform(df['LDAPS_PPT2'].values.reshape(-1,1))
df['LDAPS_PPT3'] = scaler.fit_transform(df['LDAPS_PPT3'].values.reshape(-1,1))
df['LDAPS_PPT4'] = scaler.fit_transform(df['LDAPS_PPT4'].values.reshape(-1,1))

df['DEM'] = scaler.fit_transform(df['DEM'].values.reshape(-1,1))
df['Slope'] = scaler.fit_transform(df['Slope'].values.reshape(-1,1))
```

```python
x =df.drop(columns = ['Next_Tmax','Next_Tmin'])  #features

y_max=df['Next_Tmax']  #Target column 1 "Next_Tmax"
y_min=df['Next_Tmin']  #Target column 2 "Next_Tmin"
```

```python
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
x = sc.fit_transform(x)
x
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor, GradientBoostingRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.model_selection import cross_val_score
```

```python
maxR2_Score = 0
maxRS = 0

for i in range(200):
    x_train,x_test,y_train,y_test = train_test_split(x,y_max,test_size = 0.20,random_state = i)
    LR = LinearRegression()
    LR.fit(x_train,y_train)
    predrf = LR.predict(x_test)
    Score = r2_score(y_test,predrf)
    if Score>maxR2_Score:
        maxR2_Score = Score
        maxRS = i
```
